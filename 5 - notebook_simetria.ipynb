`````python
import pandas as pd
import seaborn as sns

data = pd.read_csv('dados.csv')
`````

# Exemplo de uma população assimetrica a esquerda

```` MODA < MEDIANA < MEDIA ````

`````python
# Limitando o valor da renda até R$ 10.000
ax = sns.displot(data.query('Renda < 10000').Renda, kde=True )
ax.set_xlabels('Frequência Renda - R$')
ax.figure.set_size_inches(16, 10)
ax
`````

`````python
Moda = data.Renda.mode()
Moda
`````

`````python
Mediana = data.Renda.median()
Mediana
`````

`````python
Media = data.Renda.mean()
Media
`````

# Exemplo de população simétrica

```` MODA = MEDIANA = MEDIA ````

`````python
ax = sns.displot(data.Altura, kde=True)
ax.figure.set_size_inches(16,10)
ax.set_xlabels('Frequencia Altura - Metros')
ax
`````

`````python
Moda = data.Altura.mode()
Moda
`````

`````python
Mediana = data.Altura.median()
Mediana
`````

`````python
Media = data.Altura.mean()
Media
`````
