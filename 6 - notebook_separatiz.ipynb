```python
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv('dados.csv')
```

## Quatis, decis e percentis

```python
# Dividindo a amostragem em três quartis
data.Renda.quantile([0.25,0.5,0.75])
```

### Decis

Cria classes avaliando de 10 em 10% quantos

```python
data.Renda.quantile([i / 10 for i in range(1, 10)])
```

### Percentis

Cria classes avaliando de 1 em 1% quantos

```python
data.Renda.quantile([i / 100 for i in range(1, 100)])
```

### Histograma Acumulado

```python
ax = sns.histplot(data.Idade, cumulative=True, kde=True)
ax.figure.set_size_inches(16,10)
ax.set_title('Distribuição de Frequência Acumulada')
ax.set_ylabel('Acumulado')
ax.set_label('Anos')
ax
```

## Box-plot

```python
ax = sns.boxplot(x='Altura', data=data, orient='h')
ax.figure.set_size_inches(16,10)
ax.set_title('Altura')
ax.set_xlabel('Metros')
ax
```

### Renda média por Sexo

```python
ax = sns.boxplot(x='Renda', y='Sexo', data=data.query('Renda < 10000'), orient='h')
ax.figure.set_size_inches(16,10)
ax
```

### Escolaridade média por sexo

```python
ax = sns.boxplot(x='Anos de Estudo', y='Sexo', data=data, orient='h')
ax.figure.set_size_inches(16,10)
ax
```

```python
# UF: 29 == Bahia e UF:35 == São Paulo
ax = sns.boxplot(x = 'Renda', y = 'UF', data = data.query('(UF==35 or UF==29) and Renda < 10000'), orient = 'h')

ax.figure.set_size_inches(12, 6)
ax.set_title('Renda (R$) - Bahia X São Paulo', fontsize=18)
ax.set_xlabel('R$', fontsize=14)
ax
```
